```
#1546


N = int(input())
score = list(map(int,input().split(' ')))
M = max(score)

for i in range(0,N):
    score[i] = (score[i]/M)*100

sum = 0
for i in range(0,N):
  sum += score[i]
print(sum/N)
```

```
#8958
n = int(input())

for i in range(0,n):
  case = list(str(input()))

  score=0
  bonus_score=0
  for j in range (0,len(case)):
    if j==0:
      if case[j]=='O':
        score=1
    elif case[j]=='O':
      if case[j-1]=='O':
        bonus_score+=1
        score+=bonus_score
        score+=1
      else:
        bonus_score=0
        score+=1
    else:
      pass
  print(score) 
```
