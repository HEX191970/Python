```
#10951
import sys
while True:
    try:
        a, b = map(int, sys.stdin.readline().split())
        print(a+b)
    except:
        break;
```

```
#1110
NN = N = int(input())
count = 0

while True:
  double_digits = NN//10
  single_digit = NN%10
  result = double_digits+single_digit
  NN = (NN%10)*10+(result%10) 
  count+=1

  if (N==NN):
    print(count)
    break
```
