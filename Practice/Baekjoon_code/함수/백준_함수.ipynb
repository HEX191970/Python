```
#4673
self_number_set = set(range(1, 10001))
gener_number_set = set()

for i in range(1,10001):
  for j in str(i):
    i += int(j)
  gener_number_set.add(i)

self_number_set=self_number_set-gener_number_set

for i in sorted(self_number_set):
  print(i)
```

```
#1065 한수 
def func():
  #입력값
  N = int(input())
  #출력값
  count=0

  for i in range(1,N+1):
    if i<100:
      count+=1
    elif i<1000:
      case=[]
      for j in str(i):
        case.append(int(j))
      if case[2]+case[0]==case[1]*2:
        count+=1
    else:
      continue
  return count
```
