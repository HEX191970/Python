```
#10039
i = []

while True : 
  try : 
    a = int(input())
    if a>=40 :
      i.append(a) 
    else :
      a = 40
      i.append(a)
  except: 
    break

T = len(i)
SUM = sum(i)

print(SUM//T)
```

```
#2523
T = int(input())

for i in range(1,2*T+1):
  if i<=T:
    print('*'*i)
  else: 
    print('*'*(2*T-i))
```

```
#2446
T = int(input())

for i in range(1,T+1):
  if i<=T:
    print(' '*(i-1)+'*'*(2*T+1-2*i))
for i in range(T-1,0,-1):
  print(' '*(i-1)+'*'*(2*T+1-2*i))
```

```
#10996
T = int(input())
b, c = T,T//2
for i in range(0,T):
  if T%2==1:
    print('* '*(b-c))
    print(' *'*c)
  else:
    print('* '*c)
    print(' *'*(b-c))
```
