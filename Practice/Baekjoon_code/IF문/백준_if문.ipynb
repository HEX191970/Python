```
#9498
exam_point = int(input())

def report(a) :
  if a >= 90 : print("A")
  elif a >= 80 : print("B")
  elif a >= 70 : print("C")
  elif a >= 60 : print("D")
  else : print("F")

if exam_point <= 100 and exam_point >= 0 : report(exam_point)
else : print("오류입니다")
```

```
#2753
year = int(input())

def leap_year(a) : 
  if (a%4 == 0)==True : 
    if ((a%100 == 0)==False) or ((a%400==0)==True) : print("1")
    else : print("0")
  else : print("0")

if year>=1 and year<= 4000 :
  leap_year(year)
else : print("오류입니다")
```

```
#14681
x = int(input())
y = int(input())

def quadrant (a,b):
   if (a!=0) and (b!=0) :
     if (a>0) and (b>0) : print("1")
     elif (a<0) and (b>0) : print("2")
     elif (a<0) and (b<0) : print("3")
     else : print("4")
   else: print('사분면에 속하지 않습니다')

quadrant(x,y)
```

```
#2884
H, M = map(int, input().split(' '))

def alarm (H,M):
  if ((0<=H<24)==True) and ((0<=M<60)==True):
    if M>=45 : print(H,M-45)
    elif H>0 : print(H-1,M+15)
    else : print(23,M+15)

alarm(H,M)
```
